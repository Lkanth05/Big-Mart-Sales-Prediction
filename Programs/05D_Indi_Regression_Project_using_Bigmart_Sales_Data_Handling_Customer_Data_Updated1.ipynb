```python
train_df1 = combined_df[combined_df["train"] == 1]
test_df1 = combined_df[combined_df["test"] == 0]
train_df1.drop(["train", "test"], axis=1, inplace=True)
test_df1.drop(["test", "train", "Item_Outlet_Sales"], axis=1, inplace=True)
```

```python
train_df1.shape
```

```python
test_df1.shape
```

```python
X = train_df1.drop(['Item_Outlet_Sales'], axis = 1)
y = train_df1['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 66)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(n_estimators=300, random_state=0)
RFR.fit(X_train, y_train)
```

```python
# Calculate the accuracy of learning by the Algorithm
RFR.score(X_train,y_train)
```

```python
# Predicting the Test set results
RFR_y_pred = RFR.predict(X_test)
```

```python
RFR_Training_Acc = RFR.score(X_train,y_train)
RFR_Testing_Acc = r2_score(y_test,RFR_y_pred)
```

```python
print("Training Accuracy :", RFR_Training_Acc)
print("Testing Accuracy :", RFR_Testing_Acc)
```

```python
# Predicting the Test set results
RFR_test_pred = RFR.predict(test_df1)
```
